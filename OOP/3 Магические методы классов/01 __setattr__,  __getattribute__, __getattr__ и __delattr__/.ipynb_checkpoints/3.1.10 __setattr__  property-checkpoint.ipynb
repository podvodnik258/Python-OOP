```python



class Dimensions:
    MIN_DIMENSION = 10
    MAX_DIMENSION = 1000
    
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    @property
    def a(self):
        return self.__a
    
    @a.setter
    def a(self,value):
        self.__a = value
        
    @property
    def b(self):
        return self.__b
    
    @b.setter
    def b(self,value):
        self.__b = value
        
    @property
    def c(self):
        return self.__c
    
    @c.setter
    def c(self, value):
        if value > 0:
            self.__c = value
    
    def __setattr__(self, key, value):
        if (type(value) in (int, float) and self.MIN_DIMENSION <= value <= self.MAX_DIMENSION):
            if key in ('MIN_DIMENSION', 'MAX_DIMENSION'):
                raise AttributeError("Менять атрибуты MIN_DIMENSION и MAX_DIMENSION запрещено.")
            return super().__setattr__(key, value)
```
