```python
class DescriptorValidator:
    def __init__(self, value_type):
        self.value_type = value_type

    def __set_name__(self, owner, name):
        self.name = '_'+name

    def __set__(self, instance, value):
        if not isinstance(value, self.value_type):
            raise TypeError(f"Не верный тип переменной {self.name}. Нужен тип переменной: {self.value_type}")
        setattr(instance, self.name, value)

    def __get__(self, instance, owner):
        #print(f"call {instance.__class__.__name__} : {self.name}")
        return getattr(instance, self.name)

class Exhibit:
    name = DescriptorValidator(str)
    descr = DescriptorValidator(str)

    def __init__(self, *args, **kwargs):
        pass

    def __repr__(self):
        return f"{self.__class__.__name__}({self.name}, {self.descr})"

class Picture(Exhibit):
    author = DescriptorValidator(str)

    def __init__(self, name, author, descr):
        self.name, self.author, self.descr = (name, author, descr)

class Papyri(Exhibit):
    date = DescriptorValidator(str)

    def __init__(self, name, date, descr):
        self.name, self.date, self.descr = (name, date, descr)

class Mummies(Exhibit):
    location = DescriptorValidator(str)

    def __init__(self, name, location, descr):
        self.name, self.location, self.descr = (name, location, descr)

class Museum:
    def __init__(self, name, exhibits = []):
        self.name, self.exhibits = (name, exhibits)

    def add_exhibit(self, obj):
        self.exhibits.append(obj)

    def remove_exhibit(self, obj):
        self.exhibits.remove(obj)

    def get_info_exhibit(self, indx):
        exhibit = self.exhibits[indx]
        return f"Описание экспоната {exhibit._name}: {exhibit._descr}"
```

```python
class Artefact:
    def __init__(self, *args):
        name, another, descr = args
        self.name = name
        self.descr = descr
        if self.__class__ == "Picture":
            self.author = another
        elif self.__class__ == "Mummies":
            self.location = another
        else:
            self.date = another


class Picture (Artefact):
    pass


class Mummies (Artefact):
    pass


class Papyri (Artefact):
    pass


class Museum:
    def __init__(self, name="name"):
        self.name = name
        self.exhibits = []

    def add_exhibit(self, obj):
        self.exhibits.append(obj)

    def remove_exhibit(self, obj):
        if obj in self.exhibits: self.exhibits.remove(obj)

    def get_info_exhibit(self, indx):
        if len(self.exhibits) - 1 >= indx >= 0 or -len(self.exhibits) <= indx < 0:
            temp = self.exhibits[indx]
            return f"Описание экспоната {temp.name}: {temp.descr}"
```
