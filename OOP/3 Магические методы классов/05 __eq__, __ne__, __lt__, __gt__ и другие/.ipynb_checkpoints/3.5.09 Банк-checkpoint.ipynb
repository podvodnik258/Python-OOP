```python
class CentralBank:
    rates = {'rub': 72.5, 'dollar': 1.0, 'euro': 1.15}
    
    def __new__(cls, *args, **kwargs):
        pass
    
    @classmethod
    def register(cls, wallet):
        wallet.cb = cls
        
        
class Wallet:
    def __init__(self, volume=0):
        self.volume = volume
        self.cb = None
    
    @property
    def volume(self):
        return self.__volume
    
    @volume.setter
    def volume(self, volume):
        self.__volume = volume
        
    @property
    def cb(self):
        return self.__cb
    
    @cb.setter
    def cb(self, bank):
        self.__cb = bank
        
    def chek_bank(self):
        if not self.cb:
            raise ValueError("Неизвестен курс валют.")
            
    @staticmethod        
    def convert_rub(obj):
        rub = obj.volume * obj.cb.rates['rub'] / obj.cb.rates[obj.currency]  
        #print(f'{obj.volume} {obj.currency} = {rub} rub')
        return rub
    
    def __eq__(self, other):
        self.chek_bank()
        return self.convert_rub(self)-0.1 <= self.convert_rub(other) <= self.convert_rub(self)+0.1
    
    def __gt__(self, other):
        self.chek_bank()
        return self.convert_rub(self) > self.convert_rub(other)
    
    def __ge__(self, other):
        self.chek_bank()
        return self.convert_rub(self) >= self.convert_rub(other)

    
    
    
class MoneyR(Wallet):
    currency = 'rub'
    
class MoneyD(Wallet):
    currency = 'dollar'

class MoneyE(Wallet):
    currency = 'euro'
```

```python
CentralBank.rates = {'rub': 72.5, 'dollar': 1.0, 'euro': 1.15}

r = MoneyR(45000)
d = MoneyD(500)

CentralBank.register(r)
CentralBank.register(d)

if r > d:
    print("неплохо")
else:
    print("нужно поднажать")
```
