```python
# id в потомках
class Thing:
    ID = 0

    def __init__(self, name, price, weight = None, dims = None, memory = None, frm = None):
        self.id = self.ID
        self.name = name
        self.price = price
        self.weight = weight
        self.dims = dims
        self.memory = memory
        self.frm = frm
        
    def get_data(self):
        return (p for p in (self.id, self.name, self.price, self.weight, self.dims, self.memory, self.frm) if p)
    
class Table(Thing):
    def __init__(self, name, price, weight, dims):
        Thing.ID += 1
        super().__init__(name, price)
        self.weight = weight
        self.dims = dims
        
class ElBook(Thing):
    def __init__(self, name, price, memory, frm):
        Thing.ID += 1
        super().__init__(name, price)
        self.memory = memory
        self.frm = frm
```

```python
# HASH ID
class Thing:
    def __init__(self, name, price):
        self.id = hash(self)
        self.name = name
        self.price = price
        self.weight = self.dims = self.memory = self.frm = None

    def get_data(self):
        return self.id, self.name, self.price, self.weight, self.dims, self.memory, self.frm


class Table(Thing):
    def __init__(self, name, price, weight, dims):
        super().__init__(name, price)
        self.weight = weight
        self.dims = dims


class ElBook(Thing):
    def __init__(self, name, price, memory, frm):
        super().__init__(name, price)
        self.memory = memory
        self.frm = frm
```

```python
table = Table("Круглый", 1024, 812.55, (700, 750, 700))
book = ElBook("Python ООП", 2000, 2048, 'pdf')
print(*table.get_data())
print(*book.get_data())
```

```python
# id через метод с инкрементом
class Thing:
    __id = 0

    def __init__(self,name, price):
        self.id = self.__get_id()
        self.name = name
        self.price = price
        self.weight = None
        self.dims = None
        self.memory = None
        self.frm = None

    @classmethod
    def __get_id(cls):
        Thing.__id += 1
        return cls.__id

    def get_data(self):
        return tuple(self.__dict__.values())[1:]

class Table(Thing):
    def __init__(self, name, price, weight, dims):
        super().__init__(name, price)
        self.weight = weight
        self.dims = dims

class ElBook(Thing):
    def __init__(self, name, price, memory, frm):
        super().__init__(name, price)
        self.memory = memory
        self.frm = frm
```

```python
# id Через название класса

class Thing:
    ID = 0

    def __init__(self, name: str, price: float):
        Thing.ID += 1
        self.id = Thing.ID
        self.name = name
        self.price = price
        self.weight = None
        self.dims = None
        self.memory = None
        self.frm = None

    def get_data(self):
        return self.id, self.name, self.price, self.weight, self.dims, self.memory, self.frm


class Table(Thing):
    def __init__(self, name: str, price: float, weight: float, dims: tuple):
        Thing.__init__(self, name, price)
        self.weight = weight
        self.dims = dims


class ElBook(Thing):
    def __init__(self, name: str, price: float, memory: int, frm):
        Thing.__init__(self, name, price)
        self.memory = memory
        self.frm = frm
```

```python
# Разнообразие

class Thing:
    __id = 0

    def __init__(self, *args):
        Thing.__id += 1
        self.id = Thing.__id
        if self.__class__.__name__ == "Table":
            self.name, self.price, self.weight, self.dims = args
            self.frm = self.memory = None
        else:
            self.name, self.price, self.memory, self.frm = args
            self.weight = self.dims = None

    def get_data(self):
        return tuple(self.__dict__.values())


class Table(Thing):
    pass


class ElBook(Thing):
    pass
```
